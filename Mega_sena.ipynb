```python
# Importar as bibliotecas necessárias
import pandas as pd

# Atribuir o nome da planilha para a variável "arquivo"
arquivo = 'basedados.xlsx'

# Carregar a base de dados para um objeto pandas.io.excel.ExcelFile
excel = pd.ExcelFile(arquivo)

# Carregar a planilha excel para um objeto dataframe
dados = excel.parse(0)

# Imprimir as cinco primeiras linhas do dataframe
dados.head()
```

```python
# Diferença de valores arrecadados entre o ano de 1996 e 2017

import matplotlib.pyplot as plt

#convertendo object to datatime
dados['Data Sorteio'] = pd.to_datetime(dados['Data Sorteio'])

# Gráfico dos valores recebidos(Rateio_Sena) pelos ganhadores em 1996

# Valores dos eixos
x_values = dados['Data Sorteio'][0:41]
y_values = dados['Rateio_Sena'][0:41]

#dimensões
fig = plt.figure(figsize=(24, 12))
ax1 = fig.add_subplot(2,2,1)

#passando valores
plt.plot(x_values, y_values, c ='red')

#rotação
plt.xticks(rotation=90)

# Titulação
plt.xlabel("meses")
plt.ylabel("Valor em Milhões")
plt.title("Gráfico de 1996")


#Gráfico dos valores recebidos(Rateio_Sena) pelos ganhadores em 2017

# Valores dos eixos
a_values = dados['Data Sorteio'][1900:1997]
b_values = dados['Rateio_Sena'][1900:1997]

#dimensões
fig = plt.figure(figsize=(24, 12))
ax2 = fig.add_subplot(2,2,1)

#passando valores
ax2.plot(a_values, b_values)

#rotação
plt.xticks(rotation=90)

# Titulação
plt.xlabel("meses")
plt.ylabel("Valor em Milhões")
plt.title("Gráfico de 2017")


plt.show()

# podemos notar que temos valores em 2017 perto da casa do bilhão (potencia de 8).
# podendo assim notar que em um intervalo de 20 anos tivemos um enorme aumento na numero premiação

```

```python
# Quantidade total de Ganhadores da Sena de 1996 até 2018

#convertendo object to datatime
dados['Data Sorteio'] = pd.to_datetime(dados['Data Sorteio'])

# Valores dos eixos
x_values = dados['Data Sorteio']
y_values = dados['Ganhadores_Sena']


#dimensões
fig = plt.figure(figsize=(24, 12))

#passando valores
plt.plot(x_values, y_values, c ='red')

#rotação
plt.xticks(rotation=90)

# Titulação
plt.xlabel("Anos")
plt.ylabel("Quantidade de Ganhadores")
plt.title("Ganhadores da Sena")
```

```python
# Probabilidade do jogador ganhar na Sena, Quadra ou Quina do total de pessoas que ganharam

# Recebendo o valor de ganhadores da Sena, Quadra e Quina
G_sena = dados['Ganhadores_Sena'].sum()
G_quina = dados['Ganhadores_Quina'].sum()
G_quadra = dados['Ganhadores_Quadra'].sum()

# Calculando o valor total de ganhadores de 1996 até  início 2018
total = G_quadra + G_quina + G_sena

# Calculando as porcentagens de cada tipo
pct_sena = (G_sena * 100)/total

pct_quina = (G_quina * 100)/ total
 
pct_quadra = (G_quadra * 100)/ total

# Mostrando resultados
print("Porcentagem de você ganhar na Sena:","\n",pct_sena, "\n")

print("Porcentagem de você ganhar na Quina:","\n",pct_quina, "\n")

print("Porcentagem de você ganhar na Quadra:","\n",pct_quadra)

# Como podemos ver a porcentagem. Se você for um ganhador a probabilidade de você ter ganhado a quadra é de 98%, da quina de 1,5 
# e da Sena de 0,005 (valores arredondados).
# Conclusão: Para os ganhadores de algum prêmio a possibilidade de você estar no conjunto de ganhadores da Quadra é de quase 100%
# enquanto que ser ganhador da Sena é menor do que 1%.

```
